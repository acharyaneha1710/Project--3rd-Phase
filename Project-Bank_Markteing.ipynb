```python
df1=tin_df[['job', 'marital', 'education', 'default','housing', 'loan', 'contact', 'month', 'poutcome']]
df2=tst_df[['job', 'marital', 'education', 'default','housing', 'loan', 'contact', 'month', 'poutcome']]

df=pd.concat([df1,df2],axis=0)

x=0
for i in df:
    if x<=10:
        print('Name of the labels in the columns:',df[i].unique(),'\n',
              'No. of labels in the columns',df[i].nunique(),'\n')
    x+=1
```

# Here unknown also represent as null value. Covert all unknown value in nan value.

```python
for i in df.columns:
    tin_df[i]=tin_df[i].replace('unknown',np.nan)
    tst_df[i]=tst_df[i].replace('unknown',np.nan)

tin_df
```

# Since the dataset has a lot of string vales. We will use the encoding techniques to convert the sring data to numerical data.

```python
from sklearn.preprocessing import OrdinalEncoder
enc = OrdinalEncoder()

for i in tin_df.columns:
    if tin_df[i].dtypes == 'object':
        tin_df[i]= enc.fit_transform(tin_df[i].values.reshape(-1,1))
for i in tst_df.columns:
    if tst_df[i].dtypes == 'object':
        tst_df[i]= enc.fit_transform(tst_df[i].values.reshape(-1,1))
```

# Now Separting columns into features and target from train dataset, before transformation method

```python
x=tin_df.drop(columns=['ID','subscribed'])
y=tin_df.subscribed
```

# Try to remove skewedness using power transformation method

```python
from sklearn.preprocessing import PowerTransformer

df_pt = PowerTransformer()
df_yeo = pd.DataFrame(df_pt.fit_transform(x),columns=['age', 'job', 'marital', 'education', 'default', 'balance',
       'housing', 'loan', 'contact', 'day', 'month', 'duration', 'campaign',
       'pdays', 'previous', 'poutcome'])
df_yeo.skew().sort_values()
```

we have update our dataset with many outliers using power transformation.

# Now remove remaining outlier using Z-distribution

```python
from scipy.stats import zscore
z=np.abs(zscore(df_yeo))
threshold=3
np.where(z>3)
```

```python
df_new = pd.concat([df_yeo, y],axis=1)
df_new
df_new_z= df_new[(z<3).all(axis=1)]
df_new_z
```

```python
# Now Separting columns into features and target
x_new=df_new_z.drop(columns=['subscribed'])
y_new=df_new_z.subscribed

x_new.shape
```

# Model Development

```python
#import the necessary model algos.

#classifiaction
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier,AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB

#model selection
from sklearn.model_selection import train_test_split,cross_validate
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV

#evaluation metrics
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.metrics import roc_auc_score, roc_curve, plot_roc_curve
```

```python
LR = LogisticRegression()
DT = DecisionTreeClassifier()
KNN= KNeighborsClassifier()
RF = RandomForestClassifier()
SC = SVC()
GNB= GaussianNB()
GRA= GradientBoostingClassifier()
ADA= AdaBoostClassifier()
```

```python
model=[LogisticRegression(),DecisionTreeClassifier(),KNeighborsClassifier(),RandomForestClassifier(),
       GaussianNB(),GradientBoostingClassifier(),AdaBoostClassifier()]

models=['LR=LogisticRegression','DT=DecisionTreeClassifier','KNN=KNeighborsClassifier','RF=RandomForestClassifier',
       'GNB=GaussianNB','GRA=GradientBoostingClassifier','ADA=AdaBoostClassifier']

acc_result=[]

x_train,x_test,y_train,y_test=train_test_split(x_new,y_new,test_size=0.20)

for i in range(len(model)):
    clf=model[i]
    clf.fit(x_train,y_train)
    y_pred=clf.predict(x_test)
    acc_result.append(round (accuracy_score(y_pred,y_test)*100,2))

Result_df=pd.DataFrame({'Model Name':models, 'Accuracy Score (without tuning)':acc_result})
Result_df
```

# Logistic Regression:

```python
acc_train=[]
acc_test=[]

for i in range (0,100):
    x_train, x_test, y_train, y_test = train_test_split(x_new, y_new, test_size=0.20, random_state=i)
    
    lr = LogisticRegression()
    lr.fit(x_train,y_train)
    
    x_lr_predict=lr.predict(x_train)
    y_lr_predict=lr.predict(x_test)
    
    acc_train.append(accuracy_score(y_train,x_lr_predict))
    acc_test.append(accuracy_score(y_test,y_lr_predict))
    

acc_train_index= acc_train.index(max(acc_train))    
print("Index of max vaule of train's accuracy_score at random_state:",acc_train_index)

acc_test_index= acc_test.index(max(acc_test))
print("Index of max vaule of test's accuracy_score at random_state:",acc_test_index)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_new, y_new, test_size=0.20, random_state=89)
    
lr.fit(x_train,y_train)
x_lr_predict=lr.predict(x_train)
y_lr_predict=lr.predict(x_test)
    
print ("The training accurancy @ 89 random_state is:",accuracy_score(y_train,x_lr_predict))
print ("The testing accurancy @ 89 ranom_state is:",accuracy_score(y_test,y_lr_predict))
print ('\n Test Classification Report \n', classification_report(y_test,y_lr_predict,digits=2))
print ("\n Conconfusion_matrix:\n",confusion_matrix(y_test,y_lr_predict))
```

# Decision Tree:

```python
grid_param = {
            'criterion':['gini','entropy'],
            'max_depth': range(10,15), # the max depth of the tree.
            'min_samples_leaf': range (2,6), # the max no of samples required to be at a leaf node.
            'min_samples_split': range (3,8), # the max no of samples required to split on interval node.
            'max_leaf_nodes': range (5,10) # Best nodes are defined as a realtive in impurity. If none then nunlimited no of leaf nodes.
}
grid_search = GridSearchCV (DT, param_grid = grid_param, cv=10, n_jobs = -1)
# n_jobs : Use all the cores in your system. For performance improvement.
grid_search.fit(x_train,y_train)
```

```python
#Initiate DecisitionTreeClassifier with new parametes and train
dt=grid_search.best_estimator_

# Train the model
dt.fit(x_train,y_train)

x_dt_predict=dt.predict(x_train)
y_dt_predict=dt.predict(x_test)

print ("The training accurancy is:",accuracy_score(y_train,x_dt_predict))
print ("The testing accurancy is:",accuracy_score(y_test,y_dt_predict))
print ('\n  \n Test Classification Report \n', classification_report(y_test,y_dt_predict,digits=2))
print ("\n Conconfusion_matrix:\n",confusion_matrix(y_test,y_dt_predict))
```

# K-Nearest Neighbour:

```python
neighbors = [i for i in range(1,10)]
parameter_grid = { 'algorithm' : ['kd_tree', 'brute'],
                   'leaf_size' : [3,5,6,7,8],
                   'n_neighbors': neighbors}
gridsearch = GridSearchCV(KNN, param_grid=parameter_grid)
gridsearch.fit(x_train, y_train)
```

```python
#Initiate KNN with new parametes and train
knn=grid_search.best_estimator_

# Train the model
knn.fit(x_train,y_train)

x_knn_predict=knn.predict(x_train)
y_knn_predict=knn.predict(x_test)

print ("The training accurancy is:",accuracy_score(y_train,x_knn_predict))
print ("The testing accurancy is:",accuracy_score(y_test,y_knn_predict))
print ('\n  \n Test Classification Report \n', classification_report(y_test,y_knn_predict,digits=2))
print ("\n Conconfusion_matrix:\n",confusion_matrix(y_test,y_knn_predict))
```

# Random forest:

```python
params = {  'n_estimators' : [10,15,30,40,50],
            'criterion':['gini','entropy'],
            'max_depth': [10,15,40], # the max depth of the tree.
            'min_samples_leaf': [5,6], # the max no of samples required to be at a leaf node.
            'min_samples_split': [10,11] # the max no of samples required to split on interval node.
            #'max_leaf_nodes': range (5,10) # Best nodes are defined as a realtive in impurity. If none then nunlimited no of leaf nodes.
}

gridsearch = GridSearchCV(RF, param_grid=params,cv=10, n_jobs = -1)
gridsearch.fit(x_train,y_train)
```

```python
#Initiate RandomForestClassifier with new parametes and train
rf=gridsearch.best_estimator_

# Train the model
rf.fit(x_train,y_train)

x_rf_predict=rf.predict(x_train)
y_rf_predict=rf.predict(x_test)

print ("The training accurancy is:",accuracy_score(y_train,x_rf_predict))
print ("The testing accurancy is:",accuracy_score(y_test,y_rf_predict))
print ('\n  \n Test Classification Report \n', classification_report(y_test,y_rf_predict,digits=2))
print ("\n Conconfusion_matrix:\n",confusion_matrix(y_test,y_rf_predict))
```
