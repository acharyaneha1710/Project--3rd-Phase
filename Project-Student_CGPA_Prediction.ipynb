As per the observation all columns have grades categorical as, A+,A,A-,B+,B,B-,C+,C,C-,D+,D,F,W,WU,I.

1. "W”(Withdrawal) indicates that the student was permitted to drop the course.
    It carries no connotation of quality of student performance and is not used in the calculation of CGPA.
    that's way drop those rows which have "W" grade.

2. "WU"(Withdrawal Unauthorized) and "I"(Incomplete), progress point computation is equivalent to an “F”
    Replace all "F","WU","I" with 0 value.

3. Have to convert all Grades means categorical data to numerical data.

```python
# Drop the Withdrawal grades rows, which are not used in the calculation of CGPA.

for i in df.columns:
    df.drop(df[df.loc[:,i]=='W'].index,inplace=True)

df.reset_index(drop=True, inplace=True)
df
```

#  Replacing Grades with quivalent scale.

```python
for i in df.columns:
    df[i] = df[i].replace('A+',4.0)
    df[i] = df[i].replace('A',3.8)
    df[i] = df[i].replace('A-',3.7)
    df[i] = df[i].replace('B+',3.3)
    df[i] = df[i].replace('B',3.0)
    df[i] = df[i].replace('B-',2.7)
    df[i] = df[i].replace('C+',2.3)
    df[i] = df[i].replace('C',2.0)
    df[i] = df[i].replace('C-',1.7)
    df[i] = df[i].replace('D+',1.3)
    df[i] = df[i].replace('D',1.0)
    df[i] = df[i].replace('F',0.0)
    df[i] = df[i].replace('WU',0.0)
    df[i] = df[i].replace('I',0.0)
     
df.tail(10)
```

# Now Separting columns into features and target

```python
x=df.drop(columns=['Seat No.','CGPA'])
y=df.CGPA
```

# Now handle null values

```python
x = x.fillna(round (x.mean(),1))
x.tail(10)
```

Now the dataset is ready for make a model, bcoz we all subject has it own coorelation with target,
that's way cannot use multicorelation method. 
Also we cannot remove any skeweness or outliers form dataset, so let's further for model.

# Model Development

```python
#import the necessary model algos.
#regression
from sklearn.linear_model import LinearRegression,Ridge,Lasso,RidgeCV,LassoCV
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor,BaggingRegressor,GradientBoostingRegressor,AdaBoostRegressor
from sklearn.svm import SVR

#model selection
from sklearn.model_selection import train_test_split,cross_validate,cross_val_score
from sklearn.model_selection import GridSearchCV

#preprocessing
from sklearn.preprocessing import StandardScaler

#evaluation metrics
from sklearn.metrics import mean_squared_log_error,mean_squared_error, r2_score,mean_absolute_error # for regression
```

```python
LR = LinearRegression()
LSO= Lasso()
RIG= Ridge()
KNN= KNeighborsRegressor()
RF = RandomForestRegressor()
BAG= BaggingRegressor()
GRA= GradientBoostingRegressor()
ADA= AdaBoostRegressor()
```

```python
model=[LinearRegression(), Lasso(), Ridge(), KNeighborsRegressor(), RandomForestRegressor(), BaggingRegressor(),
       GradientBoostingRegressor(), AdaBoostRegressor()]

models=['LR = LinearRegression()','LSO = Lasso()', 'RIG = Ridge()', 'KNN = KNeighborsRegressor()',
        'RF = RandomForestRegressor()', 'BAG = BaggingRegressor()', 'GRA = GradientBoostingRegressor()',
        'ADA = AdaBoostRegressor()']

acc_result=[]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=11)

for i in range(len(model)):
    clf=model[i]
    clf.fit(x_train,y_train)
    y_pred=clf.predict(x_test)
    acc_result.append(round (r2_score(y_pred,y_test)*100,2))

Result_df=pd.DataFrame({'Model Name':models, 'Accuracy Score (without tuning)':acc_result})
Result_df
```

# Linear Regression

```python
r2s_train=[]
r2s_test=[]

for i in range (0,100):
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=i)
    
    lr = LinearRegression()
    lr.fit(x_train,y_train)
    
    x_lr_predict=lr.predict(x_train)
    y_lr_predict=lr.predict(x_test)
    
    r2s_train.append(r2_score(y_train,x_lr_predict))
    r2s_test.append(r2_score(y_test,y_lr_predict))

    
r2s_test_index= r2s_test.index(max(r2s_test))
print("Index of max vaule of test's R2 score at",r2s_test_index)
        
r2s_train_index= r2s_train.index(max(r2s_train))    
print("Index of max vaule of train's R2 score at",r2s_train_index)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=23)
    
lr.fit(x_train,y_train)
x_lr_predict=lr.predict(x_train)
y_lr_predict=lr.predict(x_test)
    
print ("The training accurancy @ 23 state is:",r2_score(y_train,x_lr_predict))
print ("The testing accurancy @ 23 state is:",r2_score(y_test,y_lr_predict))
```

```python
#Cross-Validation of the Model


Train_accuracy=r2_score(y_train,x_lr_predict)
Test_accuracy =r2_score(y_test,y_lr_predict)

from sklearn.model_selection import cross_val_score

for i in range (2,10):
    cv_score = cross_val_score(lr,x,y,cv=i)
    cv_mean = cv_score.mean()
    print("At Cross fold", i, "the cv score is", cv_mean, "\n  accuracy score for training is", Train_accuracy,
          "\n& accuracy for testing is", Test_accuracy)
    print("\n")
```
